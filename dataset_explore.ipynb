```python
%matplotlib inline
import torch
import os
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
```

```python
# Define data transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])
```

```python
# Load the DeepFish dataset
data_path = '/content/drive/MyDrive/Pytorch-Deepfish/Deep fish'  # Update this path as per your dataset location
train_dataset = datasets.ImageFolder(root=os.path.join(data_path, 'train'), transform=transform)
```

```python
# Create DataLoader instances
batch_size = 32
train_loader = DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
```

```python
# Explore sample images and labels
sample_images, sample_labels = next(iter(train_loader))
```

```python
# Display sample images (only 5 images)
num_images_to_display = 5
fig, axes = plt.subplots(1, num_images_to_display, figsize=(15, 3))

for i, (image, label) in enumerate(zip(sample_images, sample_labels)):
    if i >= num_images_to_display:
        break

    axes[i].imshow(image.permute(1, 2, 0))
    axes[i].set_title(f"Label: {label.item()}")
    axes[i].axis("off")

plt.show()
```
